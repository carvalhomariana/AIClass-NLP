```python

import nltk
from nltk.tokenize import word_tokenize
from nltk import sent_tokenize, word_tokenize, pos_tag
from nltk.corpus import stopwords
from nltk.text import Text  
from textblob import TextBlob 
```

```python
text = ("Robin Robin Robin Ben Ben Ben Miller Miller MIller Miller Miller Mariana Mariana Mariana Mariana Mariana Mariana Mariana Mariana Mariana Ciji Ciji Ciji Hasan")
```

```python
sent_tokenize = (sent_tokenize(text))
sent_tokenize
```

```python
word_token = (word_tokenize(text))
word_token
```

```python
stop_words = set(stopwords.words('english'))
stopped = [word for word in word_token if word not in stopwords.words('english')]
stopped
```

```python
from collections import Counter
c = Counter(stopped)
c.most_common()
```
